### fillna (모든 데이터 일괄로 처리)
- 수정된 결과만 알려줌
- 데이터 수정하려면 재할당

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
df1['age'] = df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
df.info()
```

```python
df1['deck'].fillna('C')
```

```python
# df['deck'].fillna('No Data')
# => 카테고리타입은 카테고리에 없는 정보는 넣을 수 없음
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치 => 남자승객의 평균
- 여자 승객의 결측치 => 여자승객의 평균

```python
df1
```

```python
cond1 = df1['sex'] == 'male'
mean_age1 = df1.loc[cond1, 'age'].mean()
df1.loc[cond1, 'age'] = df1.loc[cond1, 'age'].fillna(mean_age1)

cond2 = df1['sex'] == 'female'
mean_age2 = df1.loc[cond2, 'age'].mean()
df1.loc[cond2, 'age'] = df1.loc[cond2, 'age'].fillna(mean_age2)
df1['age'].mean()
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0]).head()
```
