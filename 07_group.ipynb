```python
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
# df1 = df.copy()
```

## apply

```python
df.head()
```

```python
df['who'].value_counts()
```

```python
def trans_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
    
```

```python
df['who'].apply(trans_who)
```

```python
df['who']
```

```python
def fare_age(x):
    return x['fare']/ x['age']
```

```python
df.apply(fare_age, axis=1)
```

```python
df['survived'].value_counts()
```

```python
df['survived'].apply(lambda x: '생존' if x == 1 else '사망')
```

## groupby

```python
df.head()
```

```python
df.groupby('sex').mean(numeric_only=True)
```

```python
df.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True)
```

```python
pd.DataFrame(df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True))
```

```python
df.groupby(['sex', 'pclass'])[['survived', 'fare']].mean(numeric_only=True)
```

## reset_index

```python
df.groupby(['sex', 'pclass'])['survived'].mean().reset_index()
```

```python
# df.groupby(['sex', 'pclass'])['survived', 'age'].mean()
# df.groupby(['sex', 'pclass'])['survived', 'age'].sum()
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean','sum'])
```

## pivot_table
어떤 데이터를 유의미한 지표가 있는지 확인 할 때

```python
df.pivot_table(index='who', values='survived')
```

```python
df.pivot_table(columns='who', values='survived')
```

```python
df.pivot_table(index=['sex', 'pclass'], values='survived')
```

```python
df.pivot_table(index=['sex', 'pclass'], values='survived', aggfunc=['sum','mean'])
```

```python
df.pivot_table(index='sex', values='pclass')
```
